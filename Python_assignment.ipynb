# Import Liabraries

```python
import numpy as np 
import pandas as pd
```

# Read_csv_file and Filtering the data_frame

```python
districts = pd.read_csv(r"C:\Users\R\Desktop\Dataset\districts.csv")
districts_filtered = districts[(districts['deleted_at'].isnull()) & (districts['status']==1)]
districts_filtered = districts_filtered[['id','name_en']]
districts_filtered = districts_filtered.rename(columns={'id':'district_id','name_en': 'district_name'})
districts_filtered.head()
```

```python
blocks = pd.read_csv(r"C:\Users\R\Desktop\Dataset\blocks.csv")
blocks_filtered = blocks[(blocks['deleted_at'].isnull()) & (blocks['status']==1)]
blocks_filtered = blocks_filtered[['id','district_id','name_en']]
blocks_filtered = blocks_filtered.rename(columns={'id':'block_id','name_en':'block_name'})
blocks_filtered.head()
```

```python
phcs = pd.read_csv(r"C:\Users\R\Desktop\Dataset\phcs.csv")
phcs_filtered = phcs[(phcs['deleted_at'].isnull()) & (phcs['status']==1)]
phcs_filtered = phcs_filtered[['id','block_id','name_en']]
phcs_filtered = phcs_filtered.rename(columns={'id':'phc_id','name_en':'phc_name'})
phcs_filtered.head()
```

```python
sub_centers = pd.read_csv(r"C:\Users\R\Desktop\Dataset\Sub_centers.csv")
sub_centers_filtered = sub_centers[(sub_centers['deleted_at'].isnull()) & (sub_centers['status']==1)]
sub_centers_filtered = sub_centers_filtered[['id','phc_id','name_en']]
sub_centers_filtered = sub_centers_filtered.rename(columns={'id':'sub_center_id','name_en':'sub_center_name'})
sub_centers_filtered.head()
```

```python
anganwadis = pd.read_csv(r"C:\Users\R\Desktop\Dataset\Anganwadis.csv")
anganwadis_filtered = anganwadis[(anganwadis['deleted_at'].isnull()) & (anganwadis['status']==1)]
anganwadis_filtered = anganwadis_filtered[['id','sub_center_id','name_en']]
anganwadis_filtered = anganwadis_filtered.rename(columns={'id':'aw_id','name_en':'aw_name'})
anganwadis_filtered.head()
```

```python
health_workers = pd.read_csv(r"C:\Users\R\Desktop\Dataset\Health_workers.csv")
health_workers_filtered = health_workers[(health_workers['deleted_at'].isnull()) & (health_workers['status']==1)]
health_workers_filtered = health_workers_filtered[['id','phc_id','sub_center_id','village_id','anganwadi_id','name','designation']]
health_workers_filtered = health_workers_filtered.rename(columns={'id':'hw_id','name':'hw_name','anganwadi_id':'aw_id'})
health_workers_filtered.head()
```

# Merging the data_frame 

```python
geo_data = pd.merge(districts_filtered,blocks_filtered,on='district_id',how='inner')
geo_data = pd.merge(geo_data,phcs_filtered,on='block_id',how='inner')
geo_data = pd.merge(geo_data,sub_centers_filtered,on='phc_id',how='inner')
geo_data = pd.merge(geo_data,anganwadis_filtered,on='sub_center_id',how='inner')
geo_data.head()
```

```python
aw_data=geo_data.copy()
asha=health_workers_filtered[['hw_id','hw_name','aw_id','designation']]
asha=health_workers_filtered[health_workers_filtered['designation']==1]
aw_data = pd.merge(aw_data,asha , on='aw_id',how='inner')
# health_workers_filtered
aw_data
```

```python
sub_center_data=geo_data.copy()
sub_center_data[['aw_id','aw_name']]=np.nan
sub_center_data=sub_center_data.drop_duplicates(subset=['sub_center_id'],keep='first')
anm=health_workers_filtered.copy()
anm=anm[['hw_id','sub_center_id','hw_name','designation']]
anm=anm[health_workers_filtered['designation']==2]
sub_center_data = pd.merge(sub_center_data,anm , on='sub_center_id',how='inner')
full_data = pd.concat([aw_data,sub_center_data],ignore_index=False)

full_data
```

```python
sub_center_data=geo_data.copy()
sub_center_data[['sub_center_id','sub_center_name','aw_id','aw_name']]=np.nan
sub_center_data=sub_center_data.drop_duplicates(subset=['phc_id'],keep='first')
mo=health_workers_filtered.copy()
mo=mo[['hw_id','phc_id','hw_name','designation']]
mo=mo[health_workers_filtered['designation']==3]
phc_data = pd.merge(sub_center_data,mo , on='phc_id',how='inner')
full_data = pd.concat([full_data,phc_data],ignore_index=False)

full_data
```

# merging the data frame full_data and sync_reports, sync_status where 1.sync_success and 2/3-sync_failed


```python
sync_reports = pd.read_csv(r"C:\Users\R\Desktop\Dataset\sync_reports.csv",low_memory='False')
sync_reports_filtered = sync_reports[(sync_reports['deleted_at'].isnull()) & (sync_reports['status']==1)]
# latest_sync=(sync_reports.groupby('entity_id').agg(latest_sync_date=('updated_at','max')))
sync_reports_filtered = sync_reports_filtered[['health_worker_id','api_log_id','entity_id','entity_name','updated_at','flag']]
sync_reports_filtered = sync_reports_filtered.rename(columns={'health_worker_id':'hw_id'})
full_data=pd.merge(full_data,sync_reports_filtered,on='hw_id',how='inner')
full_data['sync_status'] = full_data['flag'].replace({1: 'sync success', 2: 'sync failed', 3: 'sync failed'})
full_data.head()
```

# merging the data frame full_data and api_logs,  and operation api_path

```python
api_logs = pd.read_csv(r"C:\Users\R\Desktop\Dataset\api_logs.csv",low_memory='False')
api_logs_filtered= api_logs[(api_logs['deleted_at'].isnull()) & (api_logs['status']==1)]
api_logs_filtered = api_logs_filtered[['id','health_worker_id','log','api_path','created_at','updated_at','status']]
api_logs_filtered = api_logs_filtered.rename(columns={'id':'api_log_id','health_worker_id':'hw_id'})
full_data=pd.merge(full_data,api_logs_filtered,on='api_log_id',how='inner')                                    
condition=[
    full_data['api_path'].str.startswith('/chip/api/v1/sync/create/').astype(bool),
     full_data['api_path'].str.startswith('/chip/api/v1/sync/update/').astype(bool),
    full_data['api_path'].str.startswith('/chip/api/v1/sync/delete/').astype(bool)
]
values =['New record','Updated record','Deleted record']

full_data['operation']=np.select(condition,values,default='Unknown')
full_data.head()
```

# FILLING (?) VALUES 

REPLACE THE VALUE WITH MODE 

```python
df['workclass'] = df['workclass'].replace('?','private')
df['occupation'] = df['occupation'].replace('?','Prof-specialty')
df['native-country'] = df['native-country'].replace('?','United-States')
df.head(10)
```

# FEATURE ENGINEERING

```python
df.education = df.education.replace(['preschool','1st-4th','5th-6th','7th-8th','9th-10th','11th-12th'],'school')
df.education = df.education.replace('HS-grad','highschool')
df.education = df.education.replace(['some-college','Assoc-voc','Assoc-acdm','prof-school'],'higher')
df.education = df.education.replace('Bachelors','graduate')
df.education = df.education.replace('Masters','postgraduate')
df.education = df.education.replace('Doctorate','phd')
```

```python
df.education
```

```python
df['marital-status'] = df['marital-status'].replace(['Married-civ-spouse','Married-AF-spouse'],'married') 
df['marital-status'] = df['marital-status'].replace('Never-married','notmarried')
df['marital-status'] = df['marital-status'].replace(['Divorced','Separated','Widowed','Married-spouse-absent'],'other') 
df['marital-status']
```

```python
df.income = df.income.replace('<=50K',0)
df.income = df.income.replace('>50K',1)
df.income
```

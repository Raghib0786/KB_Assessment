```python
latest_sync=(sync_reports.groupby('entity_id').agg(latest_sync_date=('updated_at','max')))
```

Merged Dataframes

```python
merged_df = (
latest_sync
.merge(sync_reports_filtered, left_on=['entity_id', 'latest_sync_date'],
right_on=['entity_id', 'updated_at'], suffixes=('_sync', '_reports'))
.merge(health_workers_filtered, left_on='health_worker_id', right_on='id',
suffixes=('_reports', '_hw'))
.merge(anganwadis_filtered, left_on='anganwadi_id', right_on='id',
suffixes=('_hw', '_aw'))
.merge(sub_centers_filtered, left_on='sub_center_id_aw', right_on='id',
suffixes=('_aw', '_sub_center'))
.merge(phcs_filtered, left_on='phc_id_sub_center', right_on='id',
suffixes=('_sub_center', '_phc'))
.merge(blocks_filtered, left_on='block_id', right_on='id',
suffixes=('_phc', '_block'))
.merge(districts_filtered, left_on='district_id', right_on='id',
suffixes=('_block', '_district'))
.merge(api_logs_filtered, left_on='api_log_id', right_on='id', how='left',
suffixes=('_district', '_api_log'))
)
```

Designation_refined: 1=ASHA , 2=ANM,  3=MOIC/ Medical Officer, 4=CHA, 5=CHO, 6=OTHER


```python
merged_df['designation']=merged_df['designation'].map({
     1: 'ASHA',
    2: 'ANM',
    3: 'MOIC/ Medical Officer',
    4: 'CHA',
    5: 'CHO',
    6: 'OTHER'
})
```

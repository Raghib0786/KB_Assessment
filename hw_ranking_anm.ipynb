```python
import pandas as pd 
import numpy as np
```

```python
hw=pd.read_csv(r"C:\Users\R\Desktop\real_data\hw_master_anm.csv")
hw.head()
```

```python
hw['expected_pregnancies'] = (hw['registered_people'] * 25) / 1000
hw.head()
```

```python
# hw['% registered_people'] = (hw['registered_people'] / hw['target_pop']) * 100
# hw['% surveyed_people'] = (hw['surveyed_people'] / hw['registered_people']) * 100
hw['% expected_pregnancies'] = (hw['registered_pw'] / hw['expected_pregnancies']) * 100
hw['% early_pregnancy_registered'] = (hw['registered_within_12_weeks'] / hw['registered_pw']) * 100
hw['% atleast_4_ancs'] = (hw['atleast_4_ancs'] / hw['registered_pw']) * 100
hw['% delivery_outcome'] = (hw['total_deliveries'] / hw['registered_pw']) * 100
```

```python
indicator_columns = [
    # '% registered_people',
    # '% surveyed_people',
'% expected_pregnancies',
'% early_pregnancy_registered',
'% atleast_4_ancs',
'% delivery_outcome',
]
```

```python
hw['average_percentage'] = hw[indicator_columns].mean(axis=1)

# Rank health workers based on their average percentage
hw['rank'] = hw['average_percentage'].rank(ascending=False)

# Sort the DataFrame by rank
hw_sorted = hw.sort_values(by='rank')

# Handle infinite values
hw_sorted.replace([np.inf, -np.inf], 0, inplace=True)

# Select relevant columns to display (including original data and calculated indicators)
display_columns = hw.columns.tolist() + indicator_columns + ['average_percentage', 'rank']
```
